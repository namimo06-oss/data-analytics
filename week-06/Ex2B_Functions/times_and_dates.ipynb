Working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
print(today)
```

```python
# Format current date
formatted_today = today.strftime("%A, %B, %d, %Y")
print(f"Today is {formatted_today}")
```

# My birthday

```python
# create birthday variable
# Example: December 6, 1999

birthday = datetime.date(1999, 12, 6)
print(birthday)
```

```python
# format birthday as MM//DD/YYYY

formatted_bday = birthday.strftime("%m/%d/%Y")
print(f"My birthday is {formatted_bday}")
```

# Next 90 Days 

```python
# Add 90 days to current date 

ninety_d = today + datetime.timedelta(days=90)  # timedelta is used to add or subtract data/ it means time difference
print(ninety_d)
```

```python
# format future date
formatted_90 = ninety_d.strftime("%B %d, %Y")
print(f"90 days from today is {formatted_90}")
```

# Time for Dinner

```python
# Create a dinner time
# Example: 7:15 PM
dinner_time = datetime.time(19,15) # 19 because python uses 24-hour time
print(dinner_time)
```

```python
# format dinner time
formatted_dinner = dinner_time.strftime("%I:%M %p")
print(f"Let's meet for dinner at {formatted_dinner}")
```
