# Lab 1 - using library functions 

```python
# import random module 
# this gives access ro random function
import random
```

```python
products = [
    'Laptop', 
    'Monitor', 
    'Keyboard', 
    'Mouse', 
    'Webcam',
    'Headset', 
    'Docking Station', 
    'USB Hub', 
    'Desk Lamp', 
    'Surge Protector'
]

# product of the Day
# random.choice() selects ONE random item 

product_of_day = random.choice(products)

print (f"Product of the Day: {product_of_day}")

# select 3 products for survey 
# random.sample () selects miltiple UNIQUE items 

survey_products = random.sample(products,3)
print(f"Survey products: {survey_products}")

# shuffle product order
# random.shuffle() changes the original list 

random.shuffle(products)
print(f"Shuffle products: {products}")

# generate random transaction count
# randint includes BOTH start and end values 

transactions = random.randint(50, 300)
print(f"Today's transaction count: {transactions}")
```
