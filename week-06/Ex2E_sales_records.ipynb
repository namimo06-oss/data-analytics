This notebook analyzes a small sales dataset using Python functions.
Each function takes the full dataset as its first argument.

```python
# create sales dataset
# each tuple contains: (product, region, amount)
sales_data = [
('Laptop', 'East', 1200.00),
('Phone', 'West', 850.00),
('Tablet', 'East', 400.00),
('Monitor', 'South', 650.00),
('Keyboard', 'West', 75.00),
('Laptop', 'South', 1150.00),
('Phone', 'East', 900.00),
('Tablet', 'West', 425.00),
('Monitor', 'East', 680.00),
('Keyboard', 'South', 80.00),
('Laptop', 'West', 1300.00),
('Phone', 'West', 820.00),
]
```

## Lab 1. Total sales by region

```python
# function to calculate total sales for a specific region

def region_total(data, region):
    #sum all ammounts where region matches
    return sum(
        amt for _, r, amt in data if r == region
    )
```

```python
# test totals 
print(region_total(sales_data, 'East'))
print(region_total(sales_data, 'West'))
print(region_total(sales_data, 'South'))
```

## Lab 2. Count of sales by region

```python
# function to count records for a specific region
def region_count(data, region):
    # add 1 for every matching record
    return sum(
        1 for _, r, amt in data if r == region
    )
```

```python
# Test counts 
print(region_count(sales_data, 'East'))
print(region_count(sales_data, 'West'))
print(region_count(sales_data, 'South'))
```

## Lab 3. Average sale amount by region

```python
# function to calculate average sales for a specific region
def region_average(data, region):
    #get total sales
    total = region_total(data, region)
    #get count for records
    count = region_count(data, region)
    #prevent division by zero
    if count == 0:
        return 0

    # Calculate average 
    return total / count
```

```python
# Test averages 
print(region_average(sales_data, 'East'))
print(region_average(sales_data, 'West'))
print(region_average(sales_data, 'South'))

# Test region with no records
print(region_average(sales_data, 'North'))
```

## Lab 4. Filter by sale amount

```python
# function to return sales
# above a threshold
def sales_above(data, threshold):
    #create a list with remaining records
    return [
        record for record in data

        #record[2] = sale amount
        if record[2] > threshold
    ]
```

```python
# store results 
high_sales = sales_above(sales_data, 500)
# print each record
for record in high_sales:
    print(record)
```

```python
# BONUS LAP 4
# Function to return sales below a threshold
def sales_below(data, threshold):
    return [
        record for record in data
        if record[2] < threshold
    ]
```

```python
#store results
low_sales = sales_below(sales_data, 500)

#print each record 
for record in low_sales:
    print(record)
```

## Lab 5. Top and bottom sale

```python
# function to find highest sale
def top_sale(data):
    #compare tuples using amount
    return max(data, key=lambda s: s[2])

# function to find lowest sale
def bottom_sale(data):
    #compare tuples using amount
    return min(data, key=lambda s: s[2])
```

```python
# print highest sale
print(top_sale(sales_data))

# print lowest sale
print(bottom_sale(sales_data))
```

The top sale was a Laptop in the West region for $1300.00.

The bottom sale was a Keyboard in the West region for $75.00.

## Lab 6. Apply a discount

```python
# function to apply discount
def apply_discount(data, discount_pct):
    #use map() to update amounts
    discounted = map(
        lambda s: (
            s[0], # product
            s[1], # region
            s[2] * (1- discount_pct) #discounted amount
        ),
        data
    )

    #convert map object into list
    return list(discounted)
```

```python
# Apply 15% discount 
discounted_sales = apply_discount(sales_data, 0.15) 

# print first 3 original records
print("Original records")
for record in sales_data[:3]:
    print(record)
print()

# print first 3 discounted records
print("Discounted Records")
for record in discounted_sales[:3]:
    print(record)
```

## Lab 7. Unique regions 

```python
# function to get unique regions
def unique_regions(data):
    # extract regions and remove duplicates 
    return sorted(
        set(
            r for _, r, amt in data
        )
    )
```

```python
# print unique regions
print(unique_regions(sales_data))
```

This function could be useful in a larger analysis program because it allows the program to automically identify all regions in a dataset without manually typing them.

This pattern can also be used for finding unique products, departments, employees, categories, or customers in datasets. It helps make programs more dynamic and reusable.

## Lab 8. Region Summary

```python
# function to display region summary:
def region_summary(data, region):
    # get total sales
    total = region_total(data, region)
    # get count
    count = region_count(data, region)
    # get average 
    average = region_average(data, region)

    # narrow data to selected region
    region_data = [
        record for record in data
        if record[1] == region
    ]
    # Get top sale for that region
    top = top_sale(region_data)

    # Print formatted report
    print(f"==={region} Region ===")
    print(f"Records: {count}")
    print(f"Total Sales: ${format(total, '.2f')}")
    print(f"Average Sale: ${format(average, '.2f')}")
    print(f"Top Sale: {top[0]} - ${format(top[2], '.2f')}")
    
```

```python
# test East Region
region_summary(sales_data, 'East')
```

## Lab 9. Full summary

```python
# loop through all unique regions
for region in unique_regions(sales_data):
    # print summary for each region 
    region_summary(sales_data, region)

    print()
```
