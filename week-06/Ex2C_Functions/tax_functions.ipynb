# LAB 3 of Exercise 2C - Creating and calling functions
This lab demosnstrates how to create custom Python functions that calculate social security tax, medicare tax, and federal tax withholding based on gross pay and withholding codes.


```python
# function to calculate Social Security Tax
# Tax rate = 6.2%
def get_soc_sec_tax(gross_pay):
    #multiply gross pay by tax rate 
    soc_sec_tax = gross_pay * 0.062

    #return calculated tax 
    return soc_sec_tax
```

```python
# function to calculate medicare tax
# tax rate = 1.45%
def get_medicare_tax(gross_pay):
    #multiply gross pay by tax rate
    medicare_tax = gross_pay * 0.0145

    #return calculated tax
    return medicare_tax
```

```python
# function to calculate federal tax 
# based on withholding code
def get_federal_tax(gross_pay, withholding_code):
    # Withholding code 0
    if withholding_code == 0:
        tax_rate = 0.23
    # Withholding code 1
    elif withholding_code == 1:
        tax_rate = 0.21
    # Withholding code 2
    elif withholding_code == 2:
        tax_rate = 0.195
    # Withholding code 3
    elif withholding_code == 3:
        tax_rate = 0.185
    # Withholding code 4 or more
    else:
        # Start at 18%
        tax_rate = 0.18
        
        # Subtract 0.5% for each code over 4
        tax_rate = tax_rate - ((withholding_code - 4) * 0.005)
        
    # Calculate federal tax
    federal_tax = gross_pay * tax_rate
    # Return federal tax
    return federal_tax
```

```python
# Person 1
# Gross pay = 750
# Withholding code = 0
gross_pay = 750
withholding_code = 0
print("Person 1")
print(f"Social Security Tax: ${format(get_soc_sec_tax(gross_pay), '.2f')}")
print(f"Medicare Tax: ${format(get_medicare_tax(gross_pay), '.2f')}")
print(f"Federal Tax: ${format(get_federal_tax(gross_pay, withholding_code), '.2f')}")
```
