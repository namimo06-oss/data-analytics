Working with defining custom functions

```python
# create function with 3 parameters
# name = person's name 
# movie = favorite movie
# music = favorite musician/band 
def favorite_things(name, movie, music):
    # return formatted text
   return f""" 
My name is {name}.
My favorite moive is {movie}.
My favorite musician/band is {music}.
"""
```

```python
# call function and print result
print(favorite_things(
    "Natalia",
    "Fast and Furious",
    "Taylor Swift"
))
```

```python
# create function with one parameter
def why_im_here(purpose):
    #return purpose statement
    return purpose
```

```python
# call function and print result 
print(why_im_here(
    "I joined the Year Up United Data Analyst program to build technical skills and work toward a career at Apple or NVIDIA."
))
```

```python
# Create function with 2 parameters 
# Place = favorite place 
# why = reason why 

def favorite_place(place, why): #parameters will be place and why
    #return formatted sentence
    return f"One of my favorite places is {place} because {why}."
```

```python
# test first favorite place
print(favorite_place(
    "Miami Beach",
    "it feels peaceful and relaxing"
))
```

```python
# test second favorite place
print(favorite_place(
    "Tokyo",
    "I love the fashion, food, and technology"
))
```
