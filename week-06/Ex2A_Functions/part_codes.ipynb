# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings - part codes

```python
# Function to get supplier code
# Returns everything before the :
def get_supplier_code(part_code):

    colon_index = part_code.index(":") # find position of :

    return part_code[:colon_index] # return characters before :
```

```python
# Function to get product name
# Return everything between : and -
def get_product_num(part_code):
    
        colon_index = part_code.index(":") # find position of :

        dash_index = part_code.index("-") # find position of -

        return part_code[:colon_index + 1:dash_index] # return text between : and -
```

```python
# Function to get size
# Return everything after -
def get_size(part_code):
    
    dash_index = part_code.index("-") # find position of -

    return part_code[dash_index + 1:] # return characters after -
    
```

```python
# create variables
part1 = "ACME:123-L"  # supplied by Acme, part #123, size large
part2 = "DI:12345-M"  # supplied by DI, part #12345, size medium
part3 = "ACE:1-12"  # supplied by Ace, part #1, size 12
```

```python
# test first part code

print(f"For {part1}, the supplier is {get_supplier_code(part1)}.")

print(f"Part number {get_product_num(part1)} is available in size {get_size(part1)}.")
```

```python
# test second part code

print(f"For {part2}, the supplier is {get_supplier_code(part2)}.")

print(f"Part number {get_product_num(part2)} is available in size {get_size(part2)}.")
```

```python
# test third part code

print(f"For {part3}, the supplier is {get_supplier_code(part3)}.")

print(f"Part number {get_product_num(part3)} is available in size {get_size(part3)}.")
```

# BONUS - Using a list

```python
# Store all part codes in a list 
part_codes = [
    "ACME:123-L",
    "DI:12345-M",
    "ACE:1-12"
]

#loop through list

for code in part_codes:
    print(f"For {code}, the supplier is {get_supplier_code(code)}.")

    print(f"Part number {get_product_num(code)} is available in size {get_size(code)}.")
```
