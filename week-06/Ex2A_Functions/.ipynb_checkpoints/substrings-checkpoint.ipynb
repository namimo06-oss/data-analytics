# Exercise 2.A Using built-in functions
## Lab 1. Working with names - parsing strings

```python
test_name = 'John Doe'
```

```python
# Create a function to separate first and last name
def parse_and_display(name):
    # Loop through each character in the string 
    for i in range(len(name)):
        # check if current character is a space
        if name[i] == ' ':
            # Everything BEFORE the space becomes first name
            first = name[:i]
            # Everything AFTER the space becomes last name 
            last = name[i+1:]

            # print results
            print(f"Full name: {name}")
            print(f"First name: {first}")
            print(f"Last name: {last}")
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn'] # create a list of names
```

```python
parse_and_display(test_list[0]) # test first name on list 
```

```python
parse_and_display(test_list[1]) # test second name on list 
```

```python
parse_and_display(test_list[2]) # test third name on list 
```

```python
parse_and_display('Natalia Miranda') # test my own name 
```

```python
# Create a new function that works with 
# one word, two wordm or three-word names

def parse_and_display2(name):
    
    parts = name.split() #split the name into parts using spaces 

    if len(parts) == 1: # one word name 
        print(f"Full name: {name}")
        print(f"First name: {parts[0]}")

    elif len(parts) == 2: # two word name 
        print(f"Full name: {name}")
        print(f"First name: {parts[0]}")
        print(f"Last name: {parts[1]}")

    elif len(parts) == 3: #three word name 
        print(f"Full name: {name}")
        print(f"First name: {parts[0]}")
        print(f"Middle name: {parts[1]}")
        print(f"Last name: {parts[2]}")

    else:  # More than three words
        print("Name format not supported.")
        
```

```python
# Test one word name
parse_and_display2("Lorde")
```

```python
# Test two words name
parse_and_display2("Billie Eilish")
```

```python
#Test three words name 
parse_and_display2("Megan Thee Stallion")
```

```python
# Test four words name
parse_and_display2("Natalia Lucia Miranda Monroy")
```
