## Lab 3. Employee data 
This lab exercise makes updates to information extended extracted from an employee database

```python
# create a list of employee tuples
hr_list = [
('0123', 'HR', 'Rebecca Yang', '69000'), # index 0
('0121', 'IT', 'Mark Blick', '67000'), # index 1
('0068', 'IT', 'Kahna Larsen', '112000'), # index 2
('0234', 'OPS', 'Jim Smith', '54000') # index 3
]
```

```python
# Update Mark's last name 

# Tuples cannot be changed directly ,
# so we replace the entire tuple 
hr_list[1]= ('0121', 'IT', 'Mark Blick-Hawlkey', '67000')
```

```python
#print updated list our=tput
hr_list 
```

```python
#update Jim's department and salary 
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
```

```python
hr_list
```

```python
# print formatted employee records
for employee in hr_list:
    employee_num = employee[0]
    dept_code = employee[1]
    employee_name = employee[2]
    salary = employee[3]
    print(f"{employee_num} | {dept_code} | {employee_name} | {format(int(salary), ',')}")
```
