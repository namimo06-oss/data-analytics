# What does weights do?

In random.choices(), weights controls how likely each item is to be selected.

Each number in weights corresponds to an item in the list.
A larger weight = higher chance of being chosen.
The weights are relative, not percentages.
 # Example
import random

items = ['A', 'B', 'C']
result = random.choices(items, weights=[1, 5, 1], k=10)
print(result)
What this means:
'A' has weight 1
'B' has weight 5
'C' has weight 1

👉 'B' is more likely to appear than 'A' or 'C'.

```python
# .choices() using optional parameters
random.choices(b,weights=[10,2,1,2,1,2,1,2,6,7], k=12)   # K how many elements you want in the final list   
```


 ## Custom Functions

### working with def keywork

```python
# very simple example of a function:
# nothings happen it just create it in memory
def pizza():
    return "Yay pizza!"
```

```python
pizza()
```

```python
def pizza_v_burritos(food_name):
    if food_name == 'pizza':
        msg = 'Yay pizza!'
    elif food_name == 'burritos':
        msg = "Yay burritos!"
    else:
        msg = 'boooooo'
    return msg
```

```python
pizza_v_burritos('burritos')
```

```python
pizza_v_burritos('liver and onions')
```

```python
# pizza_v_burritos('pizza', 'burritos')
# will result in error -- function only accepts one argument 
```

```python
# example of a function with two parameters

def place_order(item, qty):
    return f'You ordered {qty} of item {item}'
```

```python
place_order('pizza', 2)
```

```python
place_order('hockey puck', 67)
```

```python
place_order(input('What would you like to order (pizza or burrito)?'), input('How many?'))
```

```python
def tax_calc(rate, amount, exempt = 'N'):
    if exempt != 'N':
        tax_amt = 0
    else:
        tax_amt = rate * amount
    return tax_amt
```

```python
tax_calc(.08, 122.45)
```

```python
tax_calc(.075, 58, 'exempt')
```

# Arbitrary Arguments 


```python
def greeting_kids(*args):
    return args
```

```python
greet = greeting_kids('Henry','Alice')

greet
```

```python
def send_greetings(whomst):
    for i in whomst:
        print(f'Hi, {i}!')
```

```python
send_greetings(greet)
```

```python
send_greetings(['Bob', 'Nancy', 'Charlemagne'])
```
