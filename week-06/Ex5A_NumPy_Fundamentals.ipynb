```python
# start by importing numpy as np
import numpy as np
```

## Array Basics

```python
# Create a normal Python list
fruits = ['apple', 'banana', 'cherry', 'date']
# print the list
print(fruits)
```

```python
# create a 1D NumPy array from fruits list
arr_1 = np.array(fruits)
# print the array
print(arr_1)
```

```python
# Check the data type of values inside arr_1
print(arr_1.dtype)

# <U6  means:
# U = unicode string
# 6 = maximum string length is 6 characters
```

```python
# create a 2D array
arr_2 = np.array([
    [10, 11, 12],
    [13, 14, 15]
])

# print the array
print(arr_2)
```

```python
# print transposed version of arr_2
print(arr_2.T)
```

```python
# Print original arr_2 again
print(arr_2)

# Using .T does NOT permanently change arr_2
# It only displays a transported version temporarily
```

```python
# Create another 2D array
arr_3 = np.array([
    [1, 3, 5],
    [2, 4, 6]
])

# print the array
print(arr_3)
```

```python
# Create arr_4 using specific values from arr_3
arr_4 = np.array([
    # Row 0, column 2 = 5
    arr_3[0, 2],
    # Row 1, column 1 = 4
    arr_3[1, 1]
])

#print arr_4
print(arr_4)
```

```python
# create a 3D array of all 1's
# with 2 values in each dimension

arr_5 = np.ones((2, 2, 2))

#print the array 
print(arr_5)
```
