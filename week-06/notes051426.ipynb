# New example: video game characters

```python
class GameCharacter: # parent class
    '''Basic model for all game characters'''
    def __init__(self, type, location):
        self.type = type
        self.location = location
    def char_profile(self):
        return f'''Character type: {self.type}
    Location: {self.location}'''
```

```python
troll = GameCharacter('troll', 'Far West')
print (troll. char_profile())
```

```python
class Enemy (GameCharacter) :
    '''Enemy troll, very nasty'''
    def __init__(self, type, location, strength):
        super().__init__(type, location)
        self.strength = strength
        self. defense = 70
    def alert(self):
        return f'Alert! You have encounteredan enemy {self.type}'
```

```python
badtroll= Enemy("Evil Troll", 'Far West', 75)
```

```python
print(badtroll.char_profile())
```

```python
print(badtroll.alert())
```
